```python
import pickle

# Convert to md
files = [
    'results_json/comve_mistral-7b-chat_10.pickle',
    'results_json/esnli_mistral-7b-chat_10.pickle'
]
for file_name in files:
    # load pickle
    with open(file_name, 'rb') as file:
        data = pickle.load(file)
        # print(data)

    results = ""
    for q in data.keys():
        results += f"### {q} \n"
        for key in data[q].keys():
            results += f"{key}: \n{data[q][key]} \n\n"
            # print(f"{key}: \n{data[q][key]} \n\n")
        # print("\pagebreak")
        # results += "\pagebreak"

    with open (file_name.split('.pickle')[0] + '.md', 'w') as file:
        file.write(results)
```
